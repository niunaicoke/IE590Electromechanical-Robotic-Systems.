```python
import sys
import time
from telemetrix import telemetrix

"""
Setup a digital pin for input pullup and monitor its changes.
"""

# global variables
BUTTON_PIN = 9  # arduino pin number (button)
LED_PIN = 6 # arduino pin number (LED)
WAIT_TIME = 2  # sleep time to keep forever loop open

# callback data indices
CB_PIN_MODE = 0
CB_PIN = 1
CB_VALUE = 2
CB_TIME = 3

def the_callback(data):
    """
    data: [pin, current reported value, pin_mode, timestamp]
    """
    date = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(data[CB_TIME]))
    print(f'Pin: {data[CB_PIN]} Value: {data[CB_VALUE]} Time Stamp: {date}')
    value = data[CB_VALUE]
    if value == 0:
        print('LED ON')
        board.digital_write(LED_PIN, 1)
    else:
        print('LED OFF')
        board.digital_write(LED_PIN, 0)
    time.sleep(WAIT_TIME)

def digital_in_pullup(my_board, pin):
    # start monitoring the pin by setting its mode
    my_board.set_pin_mode_digital_input_pullup(pin, callback=the_callback)

    # get pin changes forever
    while True:
        try:
            time.sleep(WAIT_TIME)
        except KeyboardInterrupt:
            my_board.disable_all_reporting() #stops streaming so that you do not need to re-power the board
            board.shutdown()
            sys.exit(0)


board = telemetrix.Telemetrix()
try:
    digital_in_pullup(board, BUTTON_PIN)
    board.shutdown()
except KeyboardInterrupt:
    my_board.disable_all_reporting() #stops streaming so that you do not need to re-power the board
    board.shutdown()
    sys.exit(0)
```
